```python
import pandas as pd 
import numpy as np

from scipy.sparse import csr_matrix
```

```python
df = pd.read_csv('Metrics.csv')
```

```python
df
```

```python
df.groupby('times').agg({'num_memories_recalled': 'mean'})
```

```python
data_sparse = np.ones_like(df.trial.values.astype(int))
recalls_bin = csr_matrix(
    (
        data_sparse,
        (
            df.trial.values.astype(int),
            df.times.values.astype(int),
        ),
    )
).toarray()
recalls_bincum = np.cumsum(recalls_bin, axis=1)
print(recalls_bincum)
```
